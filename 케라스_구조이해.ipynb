```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
# 구조 정의

# 1. Sequential model 만들기
model = Sequential()

# 2. layer 만들기(units = 1번째 은닉층 노드 수, input_shape=입력층 노드 수)
model.add(Dense(units=2, input_shape=(3, ), activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

```python
# 3. summary 사용하기
model.summary()
```

```python
"""
Param # 의 의미 : Weight + Bias 수
8 : Weight수 6(입력층3개x은닉층2개=6) + Bias수 2(은닉층2개=2) = 8
3 : Weight수 2(은닉층2개=2) + Bias수 1(결과층1개=1) = 3
"""
```

## 2. 케라스 compilation 이해하기

```python
# 오차(loss)를 최소화(optimizer)하고 그 정확도(accuracy)를 보여준다

from keras import optimizers
sgd = optimizers.SGD(0.1)

model.compile(loss='categorical_crossentropy', 
              optimizer=sgd,
              metrics=['accuracy'])
```
